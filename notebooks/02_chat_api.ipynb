## 4. Helper – call and print
Reusable wrapper for ad-hoc queries.

```python
import asyncio
from services.rag import chat_response

def call_chat(query: str, use_rag: bool = True, num_results: int = 5, temperature: float = 0.7, history: list = None):
    text, citations = asyncio.run(chat_response(query, use_rag=use_rag, num_results=num_results, temperature=temperature, chat_history=history or []))
    return text, citations

text, _ = call_chat("What documents do you have information about?")
print(text)
```
